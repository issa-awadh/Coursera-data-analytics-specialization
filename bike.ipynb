```python
# import relevant packages
import pandas as pd
import numpy as np
from IPython.display import display

# Importing the data
q1_19 = pd.read_csv('/content/drive/MyDrive/Coursera Capstone Project/Divvy_Trips_2019_Q1.csv')
q2_19 = pd.read_csv('/content/drive/MyDrive/Coursera Capstone Project/Divvy_Trips_2019_Q2.csv')
q3_19 = pd.read_csv('/content/drive/MyDrive/Coursera Capstone Project/Divvy_Trips_2019_Q3.csv')
q4_19 = pd.read_csv('/content/drive/MyDrive/Coursera Capstone Project/Divvy_Trips_2019_Q4.csv')
display(q1_19.head())
display(q2_19.head())
display(q3_19.head())
display(q4_19.head())
```

```python
## Import 2019 datasets
# Checking data types and inspecting the data sets
df1 = q1_19.copy()
df2 = q2_19.copy()
df3 = q3_19.copy()
df4 = q4_19.copy()
# We clean df2 columns
new_columns = list(df1.columns)
df2.columns = new_columns

display(df1.head(2))
display(df2.head(2))
display(df3.head(2))
display(df4.head(2))
```

From the inspection, we need to transform the date columns to datetime datatype and `tripduration` to float64. We also need to remove the null values in the `gender` and `birthyear` columns.<br>
We also create the column `day_of_week` but because the datetime formula give us Monday as 0 and Sunday as 7, we do further calculations to get Sunday as 1 and Saturday as 7.

```python
# Clean the tripduration column first for all
df1['tripduration'] = df1['tripduration'].str.replace(',', '')
df2['tripduration'] = df2['tripduration'].str.replace(',', '')
df3['tripduration'] = df3['tripduration'].str.replace(',', '')
df4['tripduration'] = df4['tripduration'].str.replace(',', '')

# Columns to clean datatypes to correct format
column_types = {'start_time': 'datetime64[ns]', 'end_time' : 'datetime64[ns]', 'tripduration' : float}
df1 = df1.astype(column_types)
df2 = df2.astype(column_types)
df3 = df3.astype(column_types)
df4 = df4.astype(column_types)
```

```python
# We remove null values
threshold1 = len(df1) * 0.05
cols_to_drop = df1.columns[df1.isna().sum() >= threshold1]

# We drop the null values
df1.dropna(subset=['gender', 'birthyear'], inplace=True)
df2.dropna(subset=['gender', 'birthyear'], inplace=True)
df3.dropna(subset=['gender', 'birthyear'], inplace=True)
df4.dropna(subset=['gender', 'birthyear'], inplace=True)

# We add day of week and day columns
# df1
df1['day_of_week'] = (df1['start_time'].dt.dayofweek + 1) % 7 + 1
df1['dayname_of_week'] = df1['start_time'].dt.day_name()
# df2
df2['day_of_week'] = (df2['start_time'].dt.dayofweek + 1) % 7 + 1
df2['dayname_of_week'] = df2['start_time'].dt.day_name()
# df3
df3['day_of_week'] = (df3['start_time'].dt.dayofweek + 1) % 7 + 1
df3['dayname_of_week'] = df3['start_time'].dt.day_name()
# df4
df4['day_of_week'] = (df4['start_time'].dt.dayofweek + 1) % 7 + 1
df4['dayname_of_week'] = df4['start_time'].dt.day_name()
# Display all
display(df1.head(1))
display(df2.head(1))
display(df3.head(1))
display(df4.head(1))

# Concat All
df_19 = pd.concat([df1, df2, df3, df4])
display(df_19)
```

### Customer Segmantion using clustering

```python
# -------------------------
# 4. Customer Segmentation using Clustering
# -------------------------
# For segmentation, we select features that might indicate usage patterns.
# Here we use: trip_duration, start_hour, month, and the membership flag.
features = df_19[['tripduration', 'start_hour', 'month', 'member_flag', 'age']]

# Standardize the features for clustering
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Determine the optimal number of clusters using the elbow method
inertia = []
K = range(1, 20)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 4))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters k')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Based on the elbow plot, choose an optimal number of clusters.
# For example, lets assume the elbow occurs at k = 3.
optimal_k = 7
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df_19['cluster'] = kmeans.fit_predict(features_scaled)

# -------------------------
# 5. Visualize the Clusters
# -------------------------
# Scatter plot: Trip Duration vs. Start Hour colored by cluster assignment
plt.figure(figsize=(10, 6))
sns.scatterplot(x='start_hour', y='tripduration', hue='cluster', data=df_19, palette='Set1', alpha=0.5)
plt.title("Customer Segmentation Clusters (Trip Duration vs Start Hour)")
plt.xlabel("Start Hour")
plt.ylabel("Trip Duration (minutes)")
plt.legend(title="Cluster")
plt.show()

# Pairplot: Explore pairwise relationships among features by cluster
sns.pairplot(df_19[['tripduration', 'start_hour', 'month', 'member_flag', 'cluster', 'age']],
             hue='cluster', palette='Set1', diag_kind='kde')
plt.suptitle("Pairplot of Segmentation Features", y=1.02)
plt.show()

# Summarize the cluster centers (in the scaled feature space) and also show the mean feature values per cluster
cluster_summary = df_19.groupby('cluster')[['tripduration', 'start_hour', 'month', 'member_flag', 'age']].mean()
print("Cluster Summary (Mean Values):\n", cluster_summary)

# Optionally, you can also invert the scaling of the cluster centers for interpretation:
centers_scaled = kmeans.cluster_centers_
centers = scaler.inverse_transform(centers_scaled)
centers_df = pd.DataFrame(centers, columns=features.columns)
print("Cluster Centers (Original Scale):\n", centers_df)
```
